```python
import matplotlib.pyplot as plt
import numpy as np
x1 = list(range(10))
y1 = list(range(1,20,2))
plt.scatter(x1,y1, c='#8c564b', marker=".")
plt.show()
```

Criar uma estrutura arraynumpy, ou seja, uma matriz 5x2, com os dados x e y

```python
xy = [[73,4],[62,3],[63,5],[64,56],[62,55]]
xy = np.array(xy)
# print(xy)
# print(xy[:,0],xy[:,1]) # Acessando cada coluna
# linhas, colunas = xy.media
# print(linhas)
# print(colunas)
plt.scatter(xy[:,0], xy[:,1])
plt.show()
```

```python
media = np.mean(xy,0)
print(media)
print(media[0])
print(media[1])
print(media.ndim)
std_dev = np.std(xy,0)
print(std_dev)
print(std_dev[0], std_dev[1])
```

```python
# Plotando a média dos dados

plt.scatter(xy[:,0],xy[:,1])
plt.scatter(media[0],media[1])
plt.show()
```

```python
import matplotlib.patches as patches
std_dev = np.std(xy,0) # desvio padrão de cada variável
elipse = patches.Ellipse([media[0], media[1]], std_dev[0]*2, std_dev[1]*2, alpha =0.25) # criar um elipse com o centro seja a média e largura e altura desvio padrão
fig, graph = plt.subplots()  # criar uma mesma figura para todos os gráficos
graph.scatter(xy[:,0],xy[:,1]) # gráficos dos dados
graph.scatter(media[0],media[1]) # média
graph.add_patch(elipse) # adcionando a elipse na figura criada
# graph.axis() # colocar 'equal' para ver o formato de ellipse
plt.show()
```

```python
# distância entre os pontos e a média

# dist = [round((np.linalg.norm(i-media)),4) for i in xy] # truncando casas decimais com round
dist = [np.linalg.norm(i-media) for i in xy]
print(dist)
```

```python
# normalização dos dados (com vericação)

xy_min = np.min(xy,0)
# print(xy_min)
xy_max = np.max(xy, 0)
# print(xy_max)
print(xy)
norm_xy = (xy-xy_min)/(xy_max-xy_min)
# print(xy-xy_min)
# print(xy_max-xy_min)
print(norm_xy)
# print(type(norm_xy))
```

```python
# plot com dados normalizados
media = np.mean(norm_xy,0)
std_dev = np.std(norm_xy,0) # desvio padrão de cada variável

elipse = patches.Ellipse([media[0], media[1]], std_dev[0]*2, std_dev[1]*2, alpha =0.25) # criar um elipse com o centro seja a média e largura e altura desvio padrão
fig, graph = plt.subplots()  # criar uma mesma figura para todos os gráficos

graph.scatter(norm_xy[:,0],norm_xy[:,1]) # gráficos dos dados
graph.scatter(media[0],media[1]) # média
graph.add_patch(elipse) # adcionando a elipse na figura criada
# graph.axis() # colocar 'equal' para ver o formato de ellipse
plt.show()
```
